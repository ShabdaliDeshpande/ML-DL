```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
(X_train,y_train) ,(X_test,y_test) = keras.datasets.mnist.load_data()
```

```python
len(X_train)
```

Dividing data by 255 to make it in range of 0 to 1

```python
X_train = X_train/255

X_test = X_test/255
```

```python
X_train[0].shape
```

Flattened method to assign the inputs one by one to network

```python
X_train_flattened = X_train.reshape(len(X_train),28*28)
```

```python
X_train_flattened.shape
```

```python
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
plt.matshow(X_train[0])
```

Dense(10 = output layers (0-9))

```python
model = keras.Sequential([keras.layers.Dense(10, input_shape = (784,) ,activation = 'sigmoid')])
```

```python
model.compile(optimizer ='adam',loss = 'sparse_categorical_crossentropy',metrics = ['accuracy'])
```

```python
model.fit(X_train_flattened,y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,y_test)
```

Test with flattened results

```python
predicted = model.predict(X_test_flattened)
predicted[10]
```

argmax function to show the predicted number (indices of max element in array)

```python
import numpy as np
np.argmax(predicted[10])
```
